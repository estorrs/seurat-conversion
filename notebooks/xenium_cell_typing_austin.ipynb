```python
a_rna = sc.read_h5ad(input_fp)
a_rna
```

```python
a_rna.obs
```

```python
counts = Counter(a_rna.obs[cell_type_col]).most_common()
counts
```

```python
cell_type_to_ids = {}
for cell_type in sorted(set(a_rna.obs[cell_type_col])):
    f = a_rna[a_rna.obs[cell_type_col]==cell_type]
    pool = f.obs.index.to_list()
    size = min(mapping.get(cell_type, n), f.shape[0])
    ids = np.random.choice(pool, replace=False, size=size)
    cell_type_to_ids[cell_type] = ids
    
ids = []
for ls in cell_type_to_ids.values():
    ids += ls.tolist()
filtered_rna = a_rna[ids]
```

```python
counts = Counter(filtered_rna.obs[cell_type_col]).most_common()
counts
```

```python
filtered_rna.raw.var.columns = ['index']
# filtered_rna.raw.var.index = filtered_rna.raw.var['index'].to_list()
filtered_rna.layers['counts'] = filtered_rna.raw.X

keep = ['cell_type', 'orig.ident']
filtered_rna.obs = filtered_rna.obs[keep]

filtered_rna.obs
```
