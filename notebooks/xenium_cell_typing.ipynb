```python
from collections import Counter

import scanpy as sc
import pandas as pd
import numpy as np
```

```python
n = 1000
mapping = {
    'Tumor': 5000,
}
```

## liver

```python
a_sct = sc.read_h5ad('/diskmnt/Projects/Users/estorrs/sandbox/xenium_single_cell/liver_sct.h5ad')
a_rna = sc.read_h5ad('/diskmnt/Projects/Users/estorrs/sandbox/xenium_single_cell/liver_rna.h5ad')
a_sct
```

```python
a_sct.obs
```

```python
meta = pd.read_csv(
    '/diskmnt/Projects/SenNet_analysis/Main.analysis/merged/merge_liver_combo_2/upd.cell.type/5_Merged_normalized_liver_combo.metadata_06052022.tsv',
    sep='\t', index_col=0)
meta['xenium_cell_type'] = meta['cell_type_merged'].to_list()
meta
```

```python
meta = meta[[True if x in a_sct.obs.index else False for x in meta.index]]
meta
```

```python
a_sct = a_sct[meta.index.to_list()]
a_rna = a_rna[meta.index.to_list()]

a_sct.obs['xenium_cell_type'] = meta['xenium_cell_type'].to_list()
a_rna.obs['xenium_cell_type'] = meta['xenium_cell_type'].to_list()
```

```python
counts = Counter(meta['xenium_cell_type']).most_common()
counts
```

```python
cell_type_to_ids = {}
for cell_type in sorted(set(meta['xenium_cell_type'])):
    f = a_rna[a_rna.obs['xenium_cell_type']==cell_type]
    pool = f.obs.index.to_list()
    size = min(mapping.get(cell_type, n), f.shape[0])
    ids = np.random.choice(pool, replace=False, size=size)
    cell_type_to_ids[cell_type] = ids
    
ids = []
for ls in cell_type_to_ids.values():
    ids += ls.tolist()
filtered_sct = a_sct[ids]
filtered_rna = a_rna[ids]
```

```python
counts = Counter(filtered_sct.obs['xenium_cell_type']).most_common()
counts
```

```python
np.unique(filtered_sct.X[0].toarray())
```

```python
np.unique(filtered_rna.raw.X[0].toarray())
```

```python
filtered_sct.obs = meta.loc[filtered_sct.obs.index.to_list()]

filtered_sct.raw.var.columns = ['index']
filtered_rna.raw.var.columns = ['index']
filtered_rna.raw.var.index = filtered_rna.raw.var['index'].to_list()
filtered_sct.layers['counts'] = filtered_rna.raw[:, filtered_sct.var.index.to_list()].X

keep = ['nCount_RNA', 'nFeature_RNA', 'Sample', 'Patient_ID', 'xenium_cell_type']
filtered_sct.obs = filtered_sct.obs[keep]

filtered_sct.obs
```

```python
filtered_sct.write_h5ad('/diskmnt/Projects/Users/estorrs/sandbox/xenium_single_cell/liver_downsampled.h5ad')
```
